```python
numer = input("Podaj nr tel ")
```

```python
if len(numer) == 9 and numer.isdigit():
  print("ok")

if "-" in numer:
  numer = numer.replace("-", "")
  if len(numer) == 9 and numer.isdigit():
    print("ok")

if " " in numer:
  numer = numer.replace(" ", "")
  if len(numer) == 9 and numer.isdigit():
    print("ok")
else:
  print("blad")
```
