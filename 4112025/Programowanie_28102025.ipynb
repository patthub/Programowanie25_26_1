```python
#Instrukcje warunkowe

#JEŻELI jakiś warunek jest spełniony TO:
  #wykonaj jakąś czynność

if 12 > 10:
  print("12 jest wieksze od 10")
```

```python
if 12 < 10:
  print("12 jest mniejsze od 10")
elif 12 > 10:
  print("12 jest większe od 10")
```

```python
if 12 < 10:
  print("12 jest mniejsze od 10")
elif 12 == 10:
  print("12 jest równe 10")
elif 12 <= 10:
  print("12 mniejszy lub rowny 10")
elif 12 > 10:
  print("12 wieksze od 10")
else:
  print("nic nie dziala")
```
